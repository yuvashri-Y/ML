```python
car_data['buying'],_=pd.factorize(car_data['buying'])
car_data['maint'],_=pd.factorize(car_data['maint'])
car_data['doors'],_=pd.factorize(car_data['doors'])
car_data['lug_boot'],_=pd.factorize(car_data['lug_boot'])
car_data['saftey'],_=pd.factorize(car_data['saftey'])
car_data['persons'],_=pd.factorize(car_data['persons'])
```

```python
x=car_data.iloc[:,0:-1]
y=car_data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)#train state take random numbers
```

```python
x_train
```

```python
x.columns
```

```python
from sklearn import tree,metrics
dtree=tree.DecisionTreeClassifier(criterion='gini')
dtree.fit(x_train,y_train)
```

```python
y_train_pred=dtree.predict(x_train)
```

```python
accuracy=metrics.accuracy_score(y_train,y_train_pred)
print('Train accuracy :{:.2f}'.format(accuracy))
```
