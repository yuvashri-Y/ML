```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Breast_cancer_data.csv")
```

```python
df
```

```python
df['diagnosis'].value_counts()
```

```python
x=df[['mean_radius','mean_texture','mean_perimeter','mean_area','mean_smoothness']]
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
import lightgbm as lgb
clg=lgb.LGBMClassifier()
param_grid = {
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [100, 200, 300],
    'num_leaves': [10, 20, 30],
    'max_depth': [2, 4, 6]
    # Add other hyperparameters as needed
}

```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator=clg, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)
#to avoid too many warnings
#import warnings
#warnings.filterwarning("ignore")
# Perform the grid search on the training data
grid_search.fit(x_train, y_train)
```

```python
best_params = grid_search.best_params_
print(f'Best Hyperparameters: {best_params}')
```

```python
y_pred = grid_search.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
