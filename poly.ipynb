```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv')
data
```

```python
x=data.iloc[:,1:2].values
x
```

```python
y=data.iloc[:,2:].values
y
```

```python
#model for importing linear regression
from sklearn.linear_model import LinearRegression
linear_regression=LinearRegression()
linear_regression.fit(x,y)#fit-Train
```

```python
# data->polynomial regression->linear regression->predection
from sklearn.preprocessing import PolynomialFeatures
polynomial_regression=PolynomialFeatures(degree=2)
x_poly=polynomial_regression.fit_transform(x)
linear_regression2=LinearRegression()
linear_regression2.fit(x_poly,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_regression=PolynomialFeatures(degree=3)
x_poly1=polynomial_regression.fit_transform(x)
linear_regression3=LinearRegression()
linear_regression3.fit(x_poly1,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,linear_regression.predict(x),color='blue')
plt.title('Linear regression')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,linear_regression2.predict(x_poly),color='blue')
plt.title('Linear regression2')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,linear_regression3.predict(x_poly1),color='blue')
plt.title('Linear regression 3')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,linear_regression2.predict(x_poly),color='blue')
plt.plot(x,linear_regression3.predict(x_poly1),color='green')
plt.title('Linear regression1 &2')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
x=a.drop('Y house price of unit area',axis=1)
x
```

```python
y=a['Y house price of unit area']
y
```

```python
polynomial_reg=PolynomialFeatures(degree=2,include_bias=False)
```

```python
poly_feat=polynomial_reg.fit_transform(x)
poly_feat.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(poly_feat,y,test_size=0.3,random_state=101)
```

```python
#after test_train_split -> linear regression
polymodel=LinearRegression()
```

```python
polymodel.fit(x_train,y_train)
```

```python
y_pred=polymodel.predict(x_test)
```

```python
pd.DataFrame({'Y_test': y_test,'Y_pred': y_pred,'Residuals':(y_test-y_pred)})
```

```python
poly_MAE=metrics.mean_absolute_error(y_test,y_pred)
poly_MSE=metrics.mean_squared_error(y_test,y_pred)
poly_RMSE=np.sqrt(poly_MSE)
```
